```python
#{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}
#convert to
#[Action, Adventure, Fantasy, Science Fiction]
#first see that '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#so this is string

import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
  movie_genres= []
  for i in ast.literal_eval(obj):
    movie_genres.append(i["name"])
  return movie_genres
```

```python
def convert3(obj):
  movie_cast= []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      movie_cast.append(i["name"])
      counter += 1
    else:
      break
  return movie_cast
```

```python
# One-time lookup dictionary
movie_index_map = {
    title: idx
    for idx, title in enumerate(df["title"])
}


def get_recommendation(movie, top_n=5):

    # Movie exists or not
    if movie not in movie_index_map:
        return f"Movie '{movie}' not found."

    movie_index = movie_index_map[movie]

    # Similarity scores
    distances = similarity[movie_index]

    # Sort by similarity score
    movie_list = sorted(
        enumerate(distances),
        key=lambda x: x[1],
        reverse=True
    )[1:top_n + 1]

    recommendations = []

    for idx, score in movie_list:
        recommendations.append({
            "title": df.iloc[idx]["title"],
            "similarity_score": round(float(score), 3)
        })

    return recommendations
```
